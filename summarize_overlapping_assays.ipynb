```python
import pandas as pd
import glob
import os

# Define the directory containing the CSV files
data_dir = 'data/overlapping_assays-main/datasets/source_data'

# Find all CSV files in the directory
csv_files = glob.glob(os.path.join(data_dir, '*.csv'))

print(f"Found {len(csv_files)} CSV files.")
```

```python
total_data_points = 0
file_stats = []

print("Processing files...")

for file_path in csv_files:
    try:
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        # Count rows
        num_rows = len(df)
        total_data_points += num_rows
        
        # Store stats for this file
        file_name = os.path.basename(file_path)
        file_stats.append({
            'file_name': file_name,
            'data_points': num_rows
        })
        
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Create a DataFrame for file statistics
stats_df = pd.DataFrame(file_stats)

print("Processing complete.")
```

```python
# Display Summary
print(f"Total CSV Files Processed: {len(stats_df)}")
print(f"Total Data Points (Rows): {total_data_points}")

# Show top 10 files by size
print("\nTop 10 files by number of data points:")
print(stats_df.sort_values('data_points', ascending=False).head(10))

# Show distribution statistics
print("\nDistribution of data points per file:")
print(stats_df['data_points'].describe())
```
